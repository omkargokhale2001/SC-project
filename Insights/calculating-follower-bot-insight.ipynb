```python
import pandas as pd
```

```python
labels = pd.read_csv("/kaggle/input/trimmed-twibot-22/label.csv")
```

```python
labels_dict = dict(zip(labels['id'], labels['label']))
```

```python
import csv
```

```python
interacttions = ['mentioned', 'retweeted', 'like', 'following', 'followers', 'replied_to', 'quoted']
total = 0
human_bot = 0
bot_human = 0
bot_bot = 0
human_human = 0
with open('/kaggle/input/twibot-edges/edge-009.csv', 'r') as f:
    reader = csv.reader(f)
    for i, row in enumerate(reader):
        if row[1]=="following":
            owner = row[2]
            sender = row[0]
            if labels_dict[owner]=="bot":
                if labels_dict[sender]=="bot":
                    bot_bot+=1
                elif labels_dict[sender]=="human":
                    human_bot+=1
            elif labels_dict[owner]=="human":
                if labels_dict[sender]=="bot":
                    bot_human+=1
                elif labels_dict[sender]=="human":
                    human_human+=1
            total+=1
        elif row[1]=="followers":
            owner = row[0]
            sender = row[2]
            if labels_dict[owner]=="bot":
                if labels_dict[sender]=="bot":
                    bot_bot+=1
                elif labels_dict[sender]=="human":
                    human_bot+=1
            elif labels_dict[owner]=="human":
                if labels_dict[sender]=="bot":
                    bot_human+=1
                elif labels_dict[sender]=="human":
                    human_human+=1
            total+=1
```

```python
print("bot_bot", round((bot_bot/total)*100, 2))
print("human_human", round((human_human/total)*100, 2))
print("bot_human", round((bot_human/total)*100, 2))
print("human_bot", round((human_bot/total)*100, 2))
print(total)
```
