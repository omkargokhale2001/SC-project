```python
user_dict = {"BillGates": "u50393960",
"NASA": "u11348282",
"Cristiano": "u155659213",
"CNN": "u759251",
"nytimes": "u807095",
"BarackObama": "u813286",
"justinbieber": "u27260086"
}
```

```python
def search_for_followers(user_id, row):
    if row[1]=="following" and row[2]==user_id:
        return row[0]
    elif row[1]=="followers" and row[0]==user_id:
        return row[2]
    return False
```

```python
def get_followers(user_dict):
    followers = set()
    ans = dict()
    for user, id_ in user_dict.items():
        ans[user] = set()
    with open('/kaggle/input/twibot-edges/edge-009.csv', 'r') as f:
        reader = csv.reader(f)
        for i, row in enumerate(reader):
            for username, id_ in user_dict.items():
                if search_for_followers(id_, row):
                    ans[username].add(search_for_followers(id_, row))
    return ans
    
```

```python
followers = get_followers(user_dict)
```
